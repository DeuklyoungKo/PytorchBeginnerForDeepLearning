```python
import numpy as np
from matplotlib import pyplot as plt

epoch = np.arange(1,100)
train_loss = 1.5/(epoch) #ex
val_loss = .3+1/(epoch) #ex
print("epoch\n",epoch)
print("train_loss\n",train_loss)
print("val_loss\n",val_loss)
```

```python
plt.figure(figsize=(10,5))
plt.plot(train_loss, 'r-*')
plt.plot(val_loss, 'b-*')
plt.legend(['train','validation'])
plt.title("MNIST")
plt.xlabel("eopch")
plt.ylabel("loss")
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(epoch*2,train_loss, 'r-+')
plt.plot(epoch,val_loss, 'b-*')
plt.legend(['train','validation'])
plt.title("MNIST")
plt.xlabel("eopch")
plt.ylabel("loss")
# plt.show()
plt.savefig("chap02_03_graph.png")
```
